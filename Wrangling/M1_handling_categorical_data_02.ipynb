```python
import pandas as pd
import sklearn
```

```python
from sklearn.compose import make_column_selector as selector
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_validate
```

```python
print(sklearn.__version__)
```

```python
path = "https://www.openml.org/data/get_csv/1595261/phpMawTba.csv"
adult_census = pd.read_csv(path)
adult_census.drop("fnlwgt", axis = 1, inplace = True)
```

```python
target_name = "class"
target = adult_census[target_name]
data = adult_census.drop(columns=[target_name, "education-num"])
```

```python
# Using selector to group all categorical column names
categorical_columns_selector = selector(dtype_include=object)
categorical_columns = categorical_columns_selector(data)
data_categorical = data[categorical_columns]
```

```python
print(categorical_columns)
```

```python
data_categorical["native-country"].value_counts()
```

```python
model = make_pipeline(OrdinalEncoder(handle_unknown = "use_encoded_value",
                                     unknown_value = -1),
                      LogisticRegression(max_iter = 500))
```

```python
cv_results = cross_validate(model, data_categorical, target)
cv_results
```

```python
scores = cv_results["test_score"]
print("The mean cross-validation accuracy is: "
      f"{scores.mean():.3f} +/- {scores.std():.3f}")
```

```python
# Write your code here.
model = make_pipeline(
    OneHotEncoder(handle_unknown="ignore"),
    LogisticRegression(max_iter=500))
cv_results = cross_validate(model, data_categorical, target)
scores = cv_results["test_score"]
print("The mean cross-validation accuracy is: "
      f"{scores.mean():.3f} +/- {scores.std():.3f}")
```
