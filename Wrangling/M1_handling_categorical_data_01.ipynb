```python
import pandas as pd
```

```python
from sklearn.compose import make_column_selector as selector
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.model_selection import cross_validate
```

```python
path = "https://www.openml.org/data/get_csv/1595261/phpMawTba.csv"
adult_census = pd.read_csv(path)
to_drop_list = ["fnlwgt", "education-num"]
adult_census = adult_census.drop(to_drop_list, axis = 1)
```

```python
target_name = "class"
target = adult_census[target_name]
data = adult_census.drop(columns=[target_name])
```

```python
data["native-country"].value_counts().sort_index()
```

```python
data.dtypes
```

```python
categorical_columns_selector = selector(dtype_include=object)
categorical_columns = categorical_columns_selector(data)
categorical_columns
```

```python
data_categorical = data[categorical_columns]
data_categorical.head()
```

```python
print(f"The dataset is composed of {data_categorical.shape[1]} features")
```

```python
education_column = data_categorical[["education"]]

encoder = OrdinalEncoder()
education_encoded = encoder.fit_transform(education_column)
education_encoded
```

```python
encoder.categories_
```

```python
data_encoded = encoder.fit_transform(data_categorical)
data_encoded[:5]
```

```python
encoder = OneHotEncoder(sparse_output=False)
education_encoded = encoder.fit_transform(education_column)
education_encoded
```

```python
feature_names = encoder.get_feature_names_out(input_features=["education"])
# feature_names = encoder.get_feature_names_out(input_features=["education"])
education_encoded = pd.DataFrame(education_encoded, columns=feature_names)
education_encoded[0: 5]
```

```python
print(f"The dataset is composed of {data_categorical.shape[1]} features")
data_categorical.head()
```

```python
print(f"The dataset encoded contains {data_encoded.shape[1]} features")
```

```python
print(f"The encoded dataset contains {data_encoded.shape[1]} features")
```

```python
columns_encoded = data_categorical.columns
pd.DataFrame(data_encoded, columns=columns_encoded).head()
```

```python
data["native-country"].value_counts()
```

```python
model = make_pipeline(OneHotEncoder(handle_unknown = "ignore"),
                      LogisticRegression(max_iter = 500))
```

```python
cv_results = cross_validate(model, data_categorical, target)
cv_results
```

```python
scores = cv_results["test_score"]
print(f"The accuracy is: {scores.mean():.3f} +/- {scores.std():.3f}")
```
