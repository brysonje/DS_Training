```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
path = "https://www.openml.org/data/get_csv/21854866/penguins.csv"
penguins = pd.read_csv(path).replace("?", np.nan).dropna()
```

```python
target_column = "sex"
```

```python
numerical_columns = ['culmen_length_mm', 'culmen_depth_mm']
categorical_columns = ["species"]
```

```python
for column in numerical_columns:
    penguins[column] = pd.to_numeric(penguins[column], downcast="float")
```

```python
graph1 = penguins[numerical_columns].hist(figsize=(10, 6))
```

```python
# We will plot a subset of the data to keep the plot readable and make the plotting faster
graph2 = sns.pairplot(data = penguins, vars = numerical_columns, hue = target_column,
                 plot_kws = {"alpha": 0.2}, height = 4, diag_kind = "hist", diag_kws = {"bins": 30})
```
